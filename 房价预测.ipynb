```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.sans-serif']=['SimHei'] 
plt.rcParams['axes.unicode_minus']=False

corrmat = df.corr()         
plt.subplots(figsize=(12,9))
sns.heatmap(corrmat, vmax=0.9, square=True,center=2.0)
```

```python
import numpy as np
k = 10 
plt.figure(figsize=(12,9))
cols = corrmat.nlargest(k, 'price')['price'].index
cm = np.corrcoef(df[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

```python
def spearman(df,cols):
    spr=pd.DataFrame()
    spr['feature']=cols
    spr['corr']=[df[f].corr(df['price'],'spearman')for f in cols]
    spr=spr.sort_values('corr')
    plt.figure(figsize=(6,0.3*len(cols)))
    sns.barpolt(data=spr,y='feature',x='corr',orient='h')
    
```
